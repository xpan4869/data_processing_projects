#### fix column encoding

```python
hmda.loc[:, "Census Tract Number"]
```

```python
hmda['Census Tract Number'] = hmda['Census Tract Number'].astype(str)
pd.concat([hmda['Census Tract Number'].head(5), hmda['Census Tract Number'].tail(5)])
```

#### converting to correct format

```python
import re
```

```python
# idetify possible problems
cleaned_tract_number = []
for tract in hmda['Census Tract Number']:
    if '.' not in tract:
        tract += '.'
    front, back = tract.split('.')
    front = front.zfill(4)
    back = back.ljust(2, '0')
    cleaned_tract_number.append(f'{front}.{back}')

cleaned_tract_number
```

```python
hmda['Census Tract Number'] = cleaned_tract_number
hmda['Census Tract Number']
```
